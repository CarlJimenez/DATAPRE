# Getting data directly from a website
This notebook walks you through some steps in collecting data from [Bulbapedia's National Pokedex](https://bulbapedia.bulbagarden.net/wiki/List_of_Pok%C3%A9mon_by_National_Pok%C3%A9dex_number) using `requests` and `BeautifulSoup`

### Import `requests` library
This package allows you to get any website's HTML code so that you can extract from it. Let's save the website's URL in the `URL` variable.

```python
import requests

URL="https://edition.cnn.com/world/live-news/coronavirus-pandemic-vaccine-updates-03-12-21/index.html"
```

### Load the page

```python
page=requests.get(URL)
```

```python
print(page.content)
```

### Parse HTML data

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(page.content, 'html.parser')
```

### Find all tables that contain Pokemon details

```python
# Get main content <div>
news_content=soup.find(id='posts-and-button')

# Get all <article> elements
news_articles=news_content.find_all('article')
```

### Getting List of Articles

```python
news_info=news_articles[8]
```

```python
# Check article informations (i.e. Date, Author, Title)
news_info.contents
```

```python
news_info.contents[0].text.strip()
```

```python
info_start=0

# Let's figure out how to get each item for Info
info_row=news_info.contents[info_start]

for i in range(len(info_row.contents)):
    print(f'Index {i} - {info_row.contents[i]}\n\n')

#Checking stuff
print(len(info_row.contents))
print('\n')
print(len(news_articles))
```

```python
# Extract items of interest
date_time=info_row.contents[0].text.strip()
date_time_split=date_time.split(",",1)
date=date_time_split[1]

title=info_row.contents[1].text.strip()
body=news_info.contents[1].text.strip()

if len(info_row.contents) > 2:
    author=info_row.contents[2].text.strip()
    print(f'{date}; Title: {title}; {author}\n\n')
    print(f'{body}')

else:
    print(f'{date}; Title: {title}\n\n')
    print(f'{body}')
```

### Get all Articles

```python
for i in range(info_start, len(news_articles)):
    news_info=news_articles[i]
    news_content=news_articles[i].contents
    
    info_start=0
    info_row=news_info.contents[info_start]
    
    date_time=info_row.contents[0].text.strip()
    date_time_split=date_time.split(",",1)
    date=date_time_split[1]
    
    title=info_row.contents[1].text.strip()
    body=news_info.contents[1].text.strip()

    if len(info_row.contents) > 2:
        author=info_row.contents[2].text.strip()
        print(f'{date}; Title: {title} by {author}\n')
        print(f'{body}\n\n\n')

    else:
        print(f'{date}; Title: {title}\n')
        print(f'{body}\n\n\n')
```

### Save them in a JSON

```python
import json
articles_json = []

for i in range(info_start, len(news_articles)):
    news_info=news_articles[i]
    news_content=news_articles[i].contents
    
    info_start=0
    info_row=news_info.contents[info_start]
    
    date_time=info_row.contents[0].text.strip()
    date_time_split=date_time.split(",",1)
    date=date_time_split[1]
    
    title=info_row.contents[1].text.strip()
    body=news_info.contents[1].text.strip()
    
    if len(info_row.contents) > 2:
        author=info_row.contents[2].text.strip()
        articles_json.append({
            "date": date,
            "title": title,
            "full article": body,
            "author": author
        })
    else:
        articles_json.append({
            "date": date,
            "title": title,
            "full article": body
        })
    with open('articles.json', 'w') as json_file:
        json.dump(articles_json, json_file)
    
    
articles_json
```
